```python
## Ploting the processing time
import warnings
import numpy as np
import pandas as pd
from multiprocessing import Pool, cpu_count
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc
from matplotlib import pyplot as plt
from sklearn.preprocessing import scale
import time
```

```python
import matplotlib.pyplot as plt

# Data for the first line
cores = [2, 4, 6, 8, 16, 32]
processing_timeknn = [4.99,4.05,3.43,3.01,2.84,2.66 ]

# Data for the second line
# coresRf = [2, 4, 6, 8, 16, 32]
processing_timeRf = [2.24,1.97,1.85,1.80,1.60,1.52]


# Data for the third line
# cores = [2, 4, 6, 8, 16, 32]
processing_timesvm = [7.25,7.07,6.96,6.69,6.44,6.10] 
processing_timedt= [4.75,4.44,4.27,4.18,4.10,3.98] 




# Create line plot with markers for the first line
plt.plot(cores, processing_timeknn, marker='o', color='blue', linestyle='-', linewidth=2, markersize=8)

# Create line plot with markers for the second line
plt.plot(cores, processing_timeRf, marker='s', color='green', linestyle='--', linewidth=2, markersize=8)
# Create line plot with markers for the third line
plt.plot(cores, processing_timesvm, marker='d', color='k', linestyle='--', linewidth=2, markersize=8)
# Create line plot with markers for the third line
plt.plot(cores, processing_timedt, marker='d', color='#FFA500', linestyle='--', linewidth=2, markersize=8)

# Add labels and title
plt.xlabel('Number of Cores')
plt.ylabel('Processing Time(s)')
plt.title('Processing Time vs Number of Cores')

# Add legend
plt.legend(["KNN", "RF","SVM", "DT"],loc="upper right")

# Show the plot
plt.grid(True)
plt.show()
```

### 1000 ms time window 01*700

```python
axis_font = {'fontname':'Arial', 'size':'20'}
fig = plt.figure(figsize=(8,6))
ax = fig.add_axes([0,0,1,1])
# fig, ax = plt.subplots()
Models = ['KNN','DT','XGBoost','LightGBM', 'RF', 'SVM']
accuracy_full = [96.50,98.20,99.68,99.66,99.51,97.57 ]
ax.bar(Models,accuracy_full)
ax.set_ylabel('Accuracy',**axis_font)
ax.set_xlabel('Classifier name',**axis_font)
plt.bar(Models, accuracy_full, color=['blue', 'red', 'green','black','purple','gray'])
plt.title(' 1000 ms time window',**axis_font)
plt.xticks(**axis_font)
plt.yticks(**axis_font)
# plt.rcParams.update({'font.size':30})
plt.tight_layout()
# plt.savefig('Fig_3_WholeBrain.eps')
plt.show()
```
